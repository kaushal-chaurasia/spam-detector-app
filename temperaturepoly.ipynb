```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
```

```python
temperature= [[20],[22],[24],[26],[28],[30],[32]]
sales = [10,15,25,40,60,90,130]
```

```python
df = pd.DataFrame({
    'temperature': [20, 22, 24, 26, 28, 30, 32],
    'sales': [10, 15, 25, 40, 60, 90, 130]
})
df
```

```python
X = df[['temperature']]

y= df[['sales']]
```

```python
print(X)
```

```python
print(y)
```

```python
lr= LinearRegression()
lr.fit(X,y)
```

```python
pr = PolynomialFeatures(degree = 10)
X_poly = pr.fit_transform(X)
lr_2 = LinearRegression()
lr_2.fit(X_poly, y)
```

```python
y_pred_lr = lr.predict(X)  
print(y_pred_lr)         # Linear Regression
y_pred_poly = lr_2.predict(X_poly)
print(y_pred_poly) #polynomial regression
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,y_pred_lr,color='blue')
plt.title('LinearRegression')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend(['X/y_pred_lr','X/y'],title="Sales/Level", loc='best',facecolor='white')
plt.box(False)
plt.show()
```

```python
plt.scatter(X,y,color='red')
plt.plot(X,y_pred_poly,color='blue')
plt.title('Polynomial Regression')
plt.xlabel('Temperature')
plt.ylabel('Sales')
plt.legend(['X/y_pred_lr','X/y'],title="Sales/Level", loc='best',facecolor='white')
plt.box(False)
plt.show()
```
